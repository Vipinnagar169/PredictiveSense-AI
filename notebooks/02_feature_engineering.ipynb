### Removing Useless Sensors & Preparing Data for ML

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv(r'C:\Users\vipin nagar\OneDrive\Desktop\Internship 2026\PredictiveSense-AI\data\processed\train_cleaned.csv')

print("✅ Done!")
print(f"Shape: {df.shape}")
```

```python
# Finding sensors with no variation (constant values - useless for ML)
sensor_cols = [col for col in df.columns if 'sensor' in col]

# Calculating standard deviation of each sensor
sensor_std = df[sensor_cols].std()

print("=== SENSOR STANDARD DEVIATION ===")
print(sensor_std.round(4))

# Identifying useless sensors (std = 0 or very close to 0)
useless_sensors = sensor_std[sensor_std <= 0.001].index.tolist()
print(f"\n❌ Useless Sensors (no variation): {useless_sensors}")
```

```python
# Dropping useless sensors
df = df.drop(columns=useless_sensors)

print("✅ Useless sensors dropped!")
print(f"New Shape: {df.shape}")
print(f"\nRemaining columns: {list(df.columns)}")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Selecting only sensor columns for normalization
sensor_cols = [col for col in df.columns if 'sensor' in col]

# Applying MinMaxScaler
scaler = MinMaxScaler()
df[sensor_cols] = scaler.fit_transform(df[sensor_cols])

print("✅ Data normalized successfully!")
print("\nSample normalized values:")
print(df[sensor_cols].head(3).round(4))
```

```python
import os

# Saving final ML-ready dataset
os.makedirs(r'C:\Users\vipin nagar\OneDrive\Desktop\Internship 2026\PredictiveSense-AI\data\processed', 
            exist_ok=True)

df.to_csv(r'C:\Users\vipin nagar\OneDrive\Desktop\Internship 2026\PredictiveSense-AI\data\processed\train_final.csv', 
          index=False)

print("✅ Final ML-ready dataset saved!")
print(f"Shape: {df.shape}")
print(f"\nFeatures for ML: {[col for col in df.columns if col != 'RUL']}")
print(f"\nTarget variable: RUL")
```
