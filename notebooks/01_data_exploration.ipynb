### NASA C-MAPSS Turbofan Engine Dataset

```python
# Importing all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
```

```python
# Defining column names for the dataset
column_names = ['unit_id', 'cycle', 'op_setting_1', 'op_setting_2', 'op_setting_3',
                'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5',
                'sensor_6', 'sensor_7', 'sensor_8', 'sensor_9', 'sensor_10',
                'sensor_11', 'sensor_12', 'sensor_13', 'sensor_14', 'sensor_15',
                'sensor_16', 'sensor_17', 'sensor_18', 'sensor_19', 'sensor_20',
                'sensor_21', 'sensor_22', 'sensor_23']

# Loading the training dataset
df = pd.read_csv(r'C:\Users\vipin nagar\OneDrive\Desktop\Internship 2026\PredictiveSense-AI\data\raw\train_FD001.txt', 
                 sep='\s+', 
                 header=None, 
                 names=column_names)

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
# First look at the dataset
print("=== FIRST 5 ROWS ===")
print(df.head())

print("\n=== LAST 5 ROWS ===")
print(df.tail())

print("\n=== DATASET SHAPE ===")
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
```

```python
# Detailed information about dataset
print("=== DATASET INFO ===")
print(df.info())

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())
```

```python
# Dropping columns with all missing values
df = df.drop(columns=['sensor_22', 'sensor_23'])

print("✅ Useless columns dropped!")
print(f"New Shape: {df.shape}")
print(f"Remaining Columns: {list(df.columns)}")
```

```python
# Statistical summary of the dataset
print("=== STATISTICAL SUMMARY ===")
print(df.describe().round(2))
```

```python
# Adding RUL (Remaining Useful Life) column
# RUL = max cycles of that engine - current cycle
rul = df.groupby('unit_id')['cycle'].max().reset_index()
rul.columns = ['unit_id', 'max_cycle']

df = df.merge(rul, on='unit_id')
df['RUL'] = df['max_cycle'] - df['cycle']
df = df.drop(columns=['max_cycle'])

print("✅ RUL column added successfully!")
print(f"Shape: {df.shape}")
print(df[['unit_id', 'cycle', 'RUL']].head(10))
```

```python
# Plotting RUL distribution
plt.figure(figsize=(10, 5))
plt.hist(df['RUL'], bins=50, color='steelblue', edgecolor='black')
plt.title('Distribution of RUL (Remaining Useful Life)', fontsize=14)
plt.xlabel('RUL (cycles remaining)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nMax RUL: {df['RUL'].max()}")
print(f"Min RUL: {df['RUL'].min()}")
print(f"Average RUL: {df['RUL'].mean():.2f}")
```

```python
# Plotting sensor readings for one engine over time
engine_1 = df[df['unit_id'] == 1]

plt.figure(figsize=(12, 4))
plt.plot(engine_1['cycle'], engine_1['sensor_2'], 
         color='crimson', linewidth=1.5)
plt.title('Sensor 2 Reading — Engine 1 Over Time', fontsize=14)
plt.xlabel('Cycle', fontsize=12)
plt.ylabel('Sensor 2 Value', fontsize=12)
plt.axvline(x=engine_1['cycle'].max(), 
            color='black', linestyle='--', label='Failure Point')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
